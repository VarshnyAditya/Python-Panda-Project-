```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Countries.csv')
```

```python
df
```

#Which country has the highest population

```python
df[df['population']==df['population'].max()]['country']
```

#What is the capital of Indian with highest population

```python
df[df['population']==df['population'].max()]['capital_city']
```

Which country has the least population

```python
df[df['population']==df['population'].min()]['country']
```

What is the Capital oif country with least population

```python
df[df['population']==df['population'].min()]['capital_city']
```

Top 5 countries with the highest democratic score

```python
df.nlargest(5, 'democracy_score')['country']
```

How many total region are there

```python
df['region'].value_counts().count()
```

How many times individual region has repeated

```python
df['region'].value_counts()
```

How many countries lie in eastern europe region

```python
df[df['region']=='Eastern Europe'].count()
```

Who is the leader of 2nd highest populated country

```python
df[df['population'] == df['population'].nlargest(3).iloc[1]]['political_leader']
```

How many countries are there whose political leaders are unknown

```python
df[df['political_leader'].isna()]['country'].count()
```

How many country have republic in their full name

```python
count = 0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt
```

```python
df['country_long'] = df['country_long'].apply(counting)
print(count)
```

Which country in African region has highest population

```python
africa = df[df['continent'] == 'Africa']
```

```python
africa[africa['population'] == africa['population'].max()]['country']
```
